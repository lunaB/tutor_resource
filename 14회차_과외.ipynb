```
import numpy as np
```

```
X1 = 1
X2 = 2

Y1 = 2
Y2 = 1

W11 = 0.1
W12 = 0.2
W21 = 0.3
W22 = 0.1

a = 0.1

O1 = X1 * W11 + X2 * W21
O2 = X1 * W12 + X2 * W22

def E(y, o):
    return (y - o)**2

E1 = E(Y1, O1)
E2 = E(Y2, O2)

print(O1, O2)
print(E1, E2)

# dE/dW11
# dE/dW21
dEdW11 = -2 * (Y1 - O1) * X1
dEdW21 = -2 * (Y1 - O1) * X2
W11 = W11 - a * dEdW11
W21 = W21 - a * dEdW21

# dE/dW12
# dE/dW22
dEdW12 = -2 * (Y2 - O2) * X1
dEdW22 = -2 * (Y2 - O2) * X2
W12 = W12 - a * dEdW12
W22 = W22 - a * dEdW22

print(np.array([[dEdW11, dEdW12],
                [dEdW21, dEdW22]]))

O1 = X1 * W11 + X2 * W21
O2 = X1 * W12 + X2 * W22
E1 = E(Y1, O1)
E2 = E(Y2, O2)
print(O1, O2)
print(E1, E2)
```
